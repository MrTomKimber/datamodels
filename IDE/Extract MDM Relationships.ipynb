```python
df = pd.read_csv("../sample/mdm_v11_3_pdm_additional.csv")
```

```python
df['key']=df.apply(lambda x: x['Entity'] + "." + x['Attribute'], axis=1)
tab_pk_sets_d = {k:v for k,v in df[df['IsPK']=="Yes"].pivot_table(index="Entity", values="Attribute", aggfunc=set).reset_index().values}
```

```python
single_pk_tabs_d={k:v for k,v in tab_pk_sets_d.items() if len(v)==1}
single_pk_tabs_d
```

```python
cleansed_multi_pk_tabs_d={k:v.difference({'LANG_TP_CD'}) for k,v in tab_pk_sets_d.items() if len(v)==2 and 'LANG_TP_CD' in v}
```

```python
m_use_keys = {k for k,v in single_pk_c.items() if v > 1}
s_use_keys = {k for k,v in single_pk_c.items() if v == 1}
```

```python
objs = sorted([(k,list(v)[0]) for k,v in tab_pk_sets_d.items() if any([q in v for q in m_use_keys])], key=lambda x : x[1])
```

```python
core_star_pk_sets_d={}
for k in list(m_use_keys):
    top = sorted([(a,b,difflib.SequenceMatcher(a=a,b=b).ratio()) for a,b in objs if b == k], key=lambda x : x[2])[-1]
    core_star_pk_sets_d[top[0]]=top[1]
    
core_star_pk_sets_d
    
```

```python
{k:v.difference({'LANG_TP_CD'}) for k,v in tab_pk_sets_d.items() if any([q in v for q in s_use_keys])}
```

```python
rels = []
for k,v in {k:v for k,v in tab_pk_sets_d.items() if any([q in v for q in s_use_keys])}.items():
    pk = list(v)[0]
    if pk not in {'LANG_TP_CD'}:
        to_set = set(df[(df['Attribute']==pk)&(df['IsPK']=='Yes')]['Entity'])
        from_set = set(df[(df['Attribute']==pk)&(df['IsPK']=='No')]['Entity'])
        for i in from_set:
            for j in to_set:
                rel_name =  i + " --> " + j
                from_rel = i
                to_rel = j
                from_attr = pk
                to_attr = pk
                rels.append({"Relationship" : rel_name, 
                             "FromEntity" : from_rel, "FromAttribute" : from_attr, "FromCardinality" : "",
                           "ToEntity" : to_rel, "ToAttribute" : to_attr, "Toardinality" : ""
                            })
```

```python
for k,v in {k:v for k,v in tab_pk_sets_d.items() if any([q in v for q in m_use_keys])}.items():
    pk = list(v)[0]
    if pk not in {'LANG_TP_CD'}:
        to_set = set([k for k,v in core_star_pk_sets_d.items() if v==pk])
        from_set = set(df[(df['Attribute']==pk)&(df['IsPK']=='No')]['Entity']).union(set([k for k,v in tab_pk_sets_d.items() if pk in v and k not in to_set]))
        #print(to_set)
        #print(from_set)
        for i in from_set:
            for j in to_set:
                
                rel_name =  i + " --> " + j
                from_rel = i
                to_rel = j
                from_attr = pk
                to_attr = pk
                rels.append({"Relationship" : rel_name, 
                             "FromEntity" : from_rel, "FromAttribute" : from_attr, "FromCardinality" : "",
                           "ToEntity" : to_rel, "ToAttribute" : to_attr, "Toardinality" : ""
                            })
len(rels)
```

```python
pd.DataFrame(rels).to_csv("mdm_relationships.csv")
```

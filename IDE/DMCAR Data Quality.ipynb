This might be a quick and dirty implementation in the first instance, what we want are a set of logical tests that will flag up data in the presented file that doesn't conform to an agreed specification. 

For DMCAR, this might begin with some global checks, ensuring that fieldnames are present and correct, before testing more scenario-led things like, for example, if a Relationship name is present, it should include a from and to class.

```python
from pandas import DataFrame, read_excel
import glob
import sys, os
DQ_PATH = "DQ"
import importlib



def collect_rules():
    glob_rules = glob.glob(os.path.join(DQ_PATH, "*.py"))
    for file in glob_rules:
        print(file)
        mod = file.replace("/", ".").replace(".py", "")
        modname = mod.split(".")[1]
        print(mod)
        globals()[modname]=importlib.import_module(mod)
    return {cls.name:cls() for cls in DMCAR_DQ.Function.__subclasses__()}
```

```python
dq_rules = collect_rules()
```

```python
df = read_excel("payment_example.xlsx")#.fillna("")
```

```python
rowcount=0
results = []
for i,row in df.iterrows():
    for dq, rule_class in dq_rules.items():
        if rowcount==0 and rule_class.scope == DMCAR_DQ.FunctionScope.FIRST :
            results.append((rowcount, i, dq, rule_class(row)))
        elif rule_class.scope == DMCAR_DQ.FunctionScope.ALL :
            results.append((rowcount, i, dq, rule_class(row)))
    rowcount+=1

```

```python
import math
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*math.pi*var)**.5
    num = math.exp(-(float(x)-float(mean))**2/(2*var))
    return num/denom

def box_mueller_gauss(mean=0, std=1):
    u1 = random.random()
    u2 = random.random()
    return ((sqrt(-2 * log(u1)) *sin(2 * pi * u2))*std)+mean
```

```python
def genname(forenames,sirnames):
    while not False:
        f = random.choice(forenames)
        s = random.choice(sirnames)
        yield f"{f} {s}"

def intseq(start=1000):
    while not False:
        start+=1
        yield start
        
def gendob(avg=40, std=15*365, floor=18*365):
    today = datetime.datetime.today()
    days_old = avg * 365
    while not False:
        days = box_mueller_gauss(days_old, std)
        if days > floor:
            yield (today - datetime.timedelta(days = days)).date().strftime("%Y-%m-%d")

def genchoice(choices=["Yes", "No"]):
    while not False:
        yield random.choice(choices)
```

```python
def genline(iterlist):
    while not False:
        r = ([next(i) for i in iterlist])
        yield r
```

```python
dname = genname(forenames,sirnames)
dint = intseq(start=0)
ddob = gendob()
dtype = genchoice()
iterlist=[dint,dtype,ddob,dname]
gline = genline(iterlist)
```

```python
people = []
for c in range(0,10000):
    print(next(gline))
    people.append(next(gline))
```

```python

con = sqlite3.connect("sample.db")

con.executemany("""insert into legal_entity ( le_id, le_type, le_dob , le_name) values (?,?,?,?)""", people)

con.commit()
```
